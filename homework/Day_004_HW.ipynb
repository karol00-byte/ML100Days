#### 截取部分資料

```python
df11 = app_train["SK_ID_CURR"] # select col to Series
df11
```

```python
df12=app_train[["SK_ID_CURR"]] #select col to df
df12
```

```python
df13 = app_train["SK_ID_CURR"].tolist()  #select col to list
df13
```

```python
df21=app_train["TARGET","AMT_INCOME_TOTAL"]  # 2個欄位以上要加2個括號不然會報錯 錯誤示範
df21
```

```python
df22 = app_train[["TARGET","AMT_INCOME_TOTAL"]]
df22
```

```python
df23 = app_train[["TARGET","AMT_INCOME_TOTAL"]].values.tolist()
df23
```
